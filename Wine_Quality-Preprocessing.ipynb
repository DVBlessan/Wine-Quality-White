<h1 style="font-family:newtimeroman;font-size:200%;text-align:left">Wine Quality Dataset</h1>


Wine Quality Dataset:

The Wine Quality Dataset involves predicting the quality of white wines on a scale given chemical measures of each wine.

It is a multi-class classification problem, but could also be framed as a regression problem. The number of observations for each class is not balanced. There are 4,898 observations with 11 input variables and one output variable. The variable names are as follows:

Fixed acidity,
Volatile acidity,
Citric acid,
Residual sugar,
Chlorides,
Free sulfur dioxide,
Total sulfur dioxide,
Density,
pH,
Sulphates,
Alcohol,
Quality (score between 0 and 10).

Download Source: https://machinelearningmastery.com/standard-machine-learning-datasets/

```python
#loading Dataset
wine_df=pd.read_csv("winequality-white.csv")
```

```python
#splitting the columns
wine_df[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]] = wine_df['fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"'].str.split(";",expand=True,)
```

```python
#deleting the existing columns
wine_df.drop(['fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"'],inplace=True,axis=1)
```

```python
#after splitting and deleting
#head of dataset
wine_df.head()
```

```python
#splitting Dataset
wine_df.shape
```

```python
wine_df.describe()
```

```python
wine_df.info()
```

```python
#convert object to float
columns=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']
for col in columns:
    wine_df[col]=wine_df[col].astype('float')
```

```python
#type information
wine_df.info()
```

```python
#statistical information
wine_df.describe()
```

```python
#finding missing values
wine_df.isna().sum()
```

```python
#convert response attributes to categories
values= [2, 6.5, 9]
labels = ['bad', 'good']
wine_df['quality'] = pd.cut(wine_df['quality'], bins = values, labels = labels)
wine_df.quality.value_counts()
```

```python
#handling Categories using Label Encoder
from sklearn.preprocessing import LabelEncoder
le_quality = LabelEncoder()
wine_df['quality'] = le_quality.fit_transform(wine_df['quality'])
wine_df['quality'].value_counts()
```
